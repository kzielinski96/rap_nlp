## PROCESSING DATASET

```python
import pandas as pd

songs = pd.read_csv ('dataset/main.csv')
songs.head()
```

```python
songs.info()
```

```python
num_lines = songs["number_of_positive_lines"] + songs["number_of_negative_lines"]
songs["number_of_lines"] = num_lines
pos_lines_percentage = songs["number_of_positive_lines"] / songs["number_of_lines"]
songs["postive_lines_percentage"] = pos_lines_percentage
neg_lines_percentage = songs["number_of_negative_lines"] / songs["number_of_lines"]
songs["negative_lines_percentage"] = neg_lines_percentage
```

```python
songs.head()
```

```python
dates = pd.Series(songs['song_release_date'])
dates = pd. to_datetime(dates , format = '%Y/%m/%d')
years = []

for index,i in enumerate(dates):
#     print(index, i)
    y = i.year
    years.append(y)
    
y = pd.Series(years)
songs['song_release_year'] = y
```

## REGRESSION PART

### NUMBER OF WORDS

```python
songs_ml_1 = songs_ml.drop(['number_of_positive_lines', 'number_of_negative_lines'], axis=1)
```

```python
songs_ml_1.info()
```

```python
songs_ml_1 = songs_ml_1.fillna(0)
```

```python
from sklearn.model_selection import train_test_split

songs_ml_1 = songs_ml_1.fillna(0)
x, y = songs_ml_1.loc[:, songs_ml_1.columns != 'number_of_words'], songs_ml_1.loc[:, 'number_of_words']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
y_pred = lin_reg.predict(x_test)
```

```python
import numpy as np

from sklearn.model_selection import cross_val_score

scores = cross_val_score(lin_reg, x_train, y_train,
                         scoring="neg_mean_squared_error", cv=3)
regr_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
display_scores(regr_rmse_scores)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head(25)
```

```python
from sklearn.svm import SVR

svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
svr_rbf.fit(x_train, y_train)
y_pred = svr_rbf.predict(x_test)
```

```python
scores = cross_val_score(svr_rbf, x_train, y_train,
                         scoring="neg_mean_squared_error", cv=3)
svr_rbf_rmse_scores = np.sqrt(-scores)
```

```python
display_scores(svr_rbf_rmse_scores)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head(25)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_regr = RandomForestRegressor(ccp_alpha=0.0, max_depth=9, max_features=1, max_leaf_nodes=180,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=25,
                       min_weight_fraction_leaf=0.0,
                       random_state=72)

rf_regr.fit(x_train, y_train)

y_pred = rf_regr.predict(x_test)
```

```python
scores = cross_val_score(rf_regr, x_train, y_train,
                         scoring="neg_mean_squared_error", cv=5)
rf_regr_rmse_scores = np.sqrt(-scores)
```

```python
display_scores(rf_regr_rmse_scores)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head(50)
```

## CLASSIFICATION PART

```python
from sklearn.model_selection import train_test_split

songs_ml_2 = songs_ml
songs_ml_2 = songs_ml_2.fillna(0)

x, y = songs_ml_2.loc[:, songs_ml_2.columns != 'artist_id'], songs_ml_2.loc[:, 'artist_id']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
songs_ml_2['artist_id'].value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree_clf = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint

tree_clf = DecisionTreeClassifier()

param_dist = {"max_depth": sp_randint(1,100),
              "max_features": sp_randint(0, 265),
              "min_samples_split": sp_randint(1, 100),
              "random_state": sp_randint(2, 100),
              "criterion": ["gini", "entropy"]}

# run randomized search
n_iter_search = 100
random_search = RandomizedSearchCV(tree_clf, param_distributions=param_dist,
                                   n_iter=n_iter_search, cv=5)
random_search.fit(x_train, y_train)
```

```python
print(random_search.best_score_)
print(random_search.best_params_)
```

```python
tree_clf = DecisionTreeClassifier()


tree_clf.fit(x_train,y_train)
y_pred = tree_clf.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(tree_clf, x_train, y_train, cv=3)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head(25)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)
```

```python
cross_val_score(rf, x_train, y_train, cv=3, scoring="accuracy")
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head(25)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

k_range = list(range(15, 31))
param_grid = dict(n_neighbors=k_range)
grid = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')

grid.fit(x_train, y_train)
```

```python
print(grid.best_score_)
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
knn = KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
                           metric_params=None, n_neighbors=29, p=2, weights='uniform')
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
cross_val_score(knn, x_train, y_train, cv=3, scoring="accuracy")
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head(25)
```
